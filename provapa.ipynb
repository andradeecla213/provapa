```python
#Entrada definir se é menor de idade, adulto ou idoso

idade = int( input("Digite sua idade: ") )

if idade <= 17:
  print (f"Você é menor de idade")

elif idade <= 59:
  print (f" Você é adulto")

else:
  print (f" Você é idoso")
```

```python
preço = float(input("digite o preço da compra: R$"))


desconto1 = 10
calculo1 = preço * (desconto1 / 100)
calculo1_2 = preço - calculo1

desconto2 = 20
calculo2 = preço * (desconto2 / 100)
calculo2_2 = preço - calculo2


if preço <= 100:
  print(f"sem desconto.")

elif preço >= 101 and preço <= 500:
  print(f"você tem 10% de desconto. Preço final é de: R${calculo1_2} ")

else:
  print(f"você tem 20% de desconto. Preço final é de: R${calculo2_2}")
```

```python
#triangulo

l1= int(input("digite um lado do triangulo: "))
l2= int(input("digite um lado do triangulo: "))
l3= int(input("digite um lado do triangulo: "))

if (l1+l2>l3) and (l2+l3>l1) and (l3+l1>l2):
  print("É um triangulo.")

else:
  print("Não é um triangulo")
```

```python
#ler a energia do personagem

energiap = int(input("Digite a energia do seu personagem: "))

if energiap <= 0:
  print("Personagem morto")

elif energiap >= 1 and energiap <= 30:
  print("Personagem critico")

elif energiap >= 31 and energiap <= 70:
  print("Personagem aurudo")

else:
  print("Personagem saudavel")
```

```python
#Média

nota = int (input ("Digite sua média: "))

if nota >= 7:
  print ("Você foi aprovado!!")

elif nota >= 5 and nota <= 6.9:
  print ("Você está de recuperação")

else:
  print ("Você está reprovado")
```
